### **Splitting data into Founders and Providers**

```python
founders = df[df['user_id'].str.startswith('F')]
providers = df[df['user_id'].str.startswith('S')]
```

### **Define Matching Logic & Compute Pairwise Matches**


*  **Industry match**
If startup_industry == industry_preference
Then Score = 25

*  **Skill match**
If tech_requirement == core_skill
Then score = 25

*  **Project type match**
If project_need == preferred_project_type
Then Score = 25

*  **Timeline match**
If project_deadline == availability
Then score = 25



```python
def calculate_score(founder, provider):
    industry_match = int(founder['startup_industry'] == provider['industry_preference'])
    skill_match = int(founder['tech_requirement'] == provider['core_skill'])
    timeline_match = int(founder['project_deadline'] == provider['availability'])
    project_type_match = int(founder['project_need'] == provider['preferred_project_type'])

    # Define weights
    score = (
        industry_match * 25 +
        skill_match * 25 +
        project_type_match * 25 +
        timeline_match * 25
    )

    return {
        'score': score,
        'industry_match': industry_match,
        'skill_match': skill_match,
        'project_type_match': project_type_match,
        'timeline_match': timeline_match
    }
```

### **Store and Rank Matches**

```python
matches = []

for _, f in founders.iterrows():
    for _, p in providers.iterrows():
        result = calculate_score(f, p)
        matches.append({
            'founder_id': f['user_id'],
            'provider_id': p['user_id'],
            'match_score': result['score'],
            'industry_match': result['industry_match'],
            'skill_match': result['skill_match'],
            'project_type_match': result['project_type_match'],
            'timeline_match': result['timeline_match']
        })
match_df = pd.DataFrame(matches)
```

```python
match_df
```

### **Generate Top Matches**

```python
top3_founders = match_df.groupby('founder_id').apply(lambda x: x.nlargest(3, "match_score")).reset_index(drop = True)
top3_providers = match_df.groupby('provider_id').apply(lambda x: x.nlargest(3, "match_score")).reset_index(drop = True)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_matches, x='founder_id', y='match_score', hue='provider_id')
plt.title("Top Match Score for Each Founder")
plt.xlabel("Founder ID")
plt.ylabel("Match Score")
plt.xticks(rotation=45)
plt.legend(title='Service Provider')
plt.tight_layout()
plt.show()
```
